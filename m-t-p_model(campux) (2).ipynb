```python
movies = pd.read_csv('tmdb_4900_moviessss.csv', usecols=['title','Day of the Week','Theater Type', 'ticket_price'])
```

```python
movies.head()
```

```python
movies['Theater Type'] = movies['Theater Type'].str.lower()
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies=movies[~movies['ticket_price'].isna()]
```

```python
movies.info()
```

```python
movies=movies.reset_index(drop=True)
```

```python
X =movies.drop(columns='ticket_price')
y=movies['ticket_price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['title','Day of the Week','Theater Type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['title','Day of the Week','Theater Type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
